```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

try:
    # Hapus garis miring '/' jika file diupload di folder default (current working directory) Colab
    df_kaggle = pd.read_csv('/Social_Network_Ads.csv')
    print(df_kaggle.head())

    # Memilih fitur (Umur dan Gaji) serta target (Membeli atau tidak)
    X_knn = df_kaggle[['Age', 'EstimatedSalary']]
    y_knn = df_kaggle['Purchased']

    # Normalisasi Data (SANGAT PENTING UNTUK KNN)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_knn)

    # Split data (75% training, 25% testing)
    X_train_k, X_test_k, y_train_k, y_test_k = train_test_split(X_scaled, y_knn, test_size=0.25, random_state=0)

    # Menggunakan KNN dengan K=5
    model_knn = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
    model_knn.fit(X_train_k, y_train_k)

    # Memprediksi
    y_pred_k = model_knn.predict(X_test_k)

    # Menampilkan hasil
    print("\nAkurasi KNN (K=5):", accuracy_score(y_test_k, y_pred_k))

except FileNotFoundError:
    print("Error: Silakan upload file 'Social_Network_Ads.csv' ke Colab terlebih dahulu.")
except KeyError as e:
    print(f"Error: Kolom tidak ditemukan di dataset. Pastikan nama kolom benar: {e}")
except Exception as e:
    print(f"Terjadi error lain: {e}")
```
